```python
def one_sample_z(data, mu, tails):
    n = len(data)
    xbar = np.mean(data)
    s = np.sqrt(np.sum((data-xbar)**2)/(n-1))

    zstat = (xbar-mu)/(s/np.sqrt(n))
    if tails == 'two-sided':
        p_value = 2 * (1-st.norm.cdf(abs(zstat)))
    elif tails == 'one-sided':
        if zstat < 0:
            p_value = st.norm.cdf(zstat) # lower tail by default
        else:
            p_value = 1-st.norm.cdf(zstat)
    else:
        raise ValueError("Invalid value for 'tails' parameter")
    
    return zstat, p_value
```

```python
one_sample_z(data, 100, 'two-sided')
```

```python
def one_sample_t(data, mu, tails):
    n = len(data)
    xbar = np.mean(data)
    s = np.sqrt(np.sum((data-xbar)**2)/(n-1))
    tstat = (xbar-mu)/(s/np.sqrt(n))
    df = n-1
    if tails == 'two-sided':
        p_value = 2 * (1-st.t.cdf(abs(tstat),df))
    elif tails == 'one-sided':
        if tstat < 0:
            p_value = st.t.cdf(tstat, df)
        else:
            p_value = 1 - st.t.cdf(tstat, df)
    else:
        raise ValueError("Invalid values for 'tails' parameter.")
    
    return tstat, p_value
```

```python
one_sample_t(sample,pop_mean,'two-sided')
```

# Paired T-Test

```python
sample1 = np.array([10,12,14,15,16,17,18,19,20])
sample2 = np.arange(12,30,2)
```

```python
len(sample1)
```

```python
def paired_ttest(sample1, sample2, tails):
    n1 = len(sample1)
    n2 = len(sample2)
    if n1 == n2:
        d = sample1-sample2
        dbar = np.mean(d)
        sd = np.sqrt(np.sum((d-dbar)**2)/(n1-1))
        tstat = dbar/(sd/np.sqrt(n1))
        df = n1-1
        if tails == 'two-sided':
            p_value = 2 * (1-st.t.cdf(abs(tstat),df))
        elif tails == 'one-sided':
            if tstat < 0:
                p_value = st.t.cdf(tstat, df)
            else:
                p_value = 1 - st.t.cdf(tstat, df)
        else:
           raise ValueError("Invalid values for 'tails' parameter.")
    else:
        raise ValueError("Different length of data")

    return tstat, p_value 
        
```

```python
paired_ttest(sample1,sample2,'one-sided')
```

# Independent T-Test

```python
data1 = np.array([10,12,14,15,16])
data2 = np.array([13,15,17,19,21])
```

```python
np.var(data1)
```

```python
def independent_ttest(input1, input2, tails, equal_var):
    n1 = len(input1)
    n2 = len(input2)
    xbar1 = np.mean(input1)
    xbar2 = np.mean(input2)
    s1 = np.var(input1,ddof=1)
    s2 = np.var(input2,ddof=1)

    if equal_var:
        df = n1+n2-2
        s_pooled = np.sqrt((((n1-1)*s1) + ((n2-1)*s2))/df)
        tstat = (xbar1 - xbar2)/(s_pooled * np.sqrt((1/n1)+(1/n2)))
    else:
        df = ((s1/n1 + s2/n2)**2)/(((s1/n1)**2/(n1-1))+((s2/n2)**2/(n2-1)))
        tstat = (xbar1 - xbar2)/np.sqrt((s1/n1) + (s2/n2))

    if tails == 'two-sided':
        p_value = 2 * (1-st.t.cdf(abs(tstat),df))
    elif tails == 'one-sided':
        if tstat < 0:
            p_value = st.t.cdf(tstat, df)
        else:
            p_value = 1 - st.t.cdf(tstat, df)
    else:
        raise ValueError("Invalid values for 'tails' parameter.")
    
    return tstat, df, p_value
    
```

```python
independent_ttest(data1,data2,'two-sided',equal_var=False)
```

# 1 Sample Proportion

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
def prop_test(x, n, p0, tails):
    p = x/n
    q0 = 1-p0
    zstat = (p-p0)/np.sqrt(p0*q0/n)
    if tails == 'two-sided':
        p_value = 2 * (1-st.norm.cdf(abs(zstat)))
    elif tails == 'one-sided':
        if tstat < 0:
            p_value = st.norm.cdf(zstat)
        else:
            p_value = 1 - st.norm.cdf(zstat)
    else:
        raise ValueError("Invalid values for 'tails' parameter.")
    
    return zstat, p_value
```

```python
import statsmodels.api as sm

# Load the data set as a dataframe-of-dataframes
data = sm.datasets.spector.load_pandas()
# Extract the complete dataset
df = data['data']
# Rename
df = df.rename(columns={
    'TUCE': 'test_score',
    'PSI': 'participated',
    'GRADE': 'grade_improved'
})

print(df[15:22])
```

```python
# Number of successes
count = len(df[df['grade_improved'] == 1])
# Number of observations
nobs = len(df)
# Proportion under the null hypothesis
pi_0 = 0.5
# Perform a one-sample Z-test for a proportion
z_stat, p_value = proportions_ztest(count, nobs, value=pi_0, prop_var=pi_0)
# Proportion successful
pi = count / nobs

print(f'Proportion,  = {pi:.1%}; Z-statistic = {z_stat:.3f}; p = {p_value:.3f}')
```

# 2 Sample Proportion Test

```python
def two_sample_proptest(x1,x2,n1,n2,tails):
    p1 = x1/n1
    p2 = x2/n2
    p_hat = (x1+x2) / (n1+n2)
    q_hat = 1-p_hat

    zstat = (p1-p2) / np.sqrt((p_hat * q_hat * ((1/n1)+(1/n2))))
    if tails == 'two-sided':
        p_value = 2 * (1-st.norm.cdf(abs(zstat)))
    elif tails == 'one-sided':
        if tstat < 0:
            p_value = st.norm.cdf(zstat)
        else:
            p_value = 1 - st.norm.cdf(zstat)
    else:
        raise ValueError("Invalid values for 'tails' parameter.")
    
    return zstat, p_value
```

```python
two_sample_proptest(31,19,50,50,'two-sided')
```

```python
count = np.array([31, 19])
nobs = np.array([50, 50])

zscore, pval = proportions_ztest(count, nobs, alternative='two-sided')
```

```python
import pandas as pd
```

```python
placebo = [3,2,1,1,4]
low = [5,2,4,2,3]
high = [7,4,5,3,6]
```

```python
data = []
data.append(placebo)
data.append(low)
data.append(high)


data
```

```python
df = pd.DataFrame(data).transpose()
df.columns = ['placebo','low','high']
df
```

```python
from scipy.stats import f
```

```python
def anova(data):
    nk = len(data.columns)
    nt = len(data)
    xbar = data.mean()
    xbar_tot = np.mean(data.values)
    ssw = 0
    sst = 0
    nT = 0

    for i in range(nt):
        nT += len(data.iloc[i])

    for i in range(nt):
        for j in range(nk):
            ssw += (data.iloc[i,j] - xbar.iloc[j])**2
            sst += (data.iloc[i,j] - xbar_tot)**2
    
    ssb = sst - ssw
    
    msb = ssb / (nk-1)
    msw = ssw / (nT-nk)

    fstat = msb/msw

    dfb = nk - 1
    dfw = nT - nk
    p_value = 1 - f.cdf(fstat, dfb, dfw)

    return fstat, ssb, ssw, msb, msw, p_value
```
